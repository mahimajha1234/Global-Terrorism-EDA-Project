# Import all libraries:-

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

# Import the dataset:-

```python
df = pd.read_csv("Terrorism.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
for i in df.columns:
    print(i,end = ",")
```

```python
df = df[["iyear","imonth","iday","country_txt","region_txt","provstate","city","latitude","longitude","location","summary",
         "attacktype1_txt","targtype1_txt","gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
df.head()
```

```python
df.rename(columns = {"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","region_txt":"Region",
          "provstate":"Province/State","city":"City","latitude":"Latitude","longitude":"Longitude","location":"Location",
          "summary":"summary","attacktype1_txt": "Attack Type","tagtype1_txt":"Target Type","gname":"Group name",
          "motive":"Motive","weaptype1_txt":"Weapon Type","nkill":"Killed","nwound":"Wounded","addnotes":"Add notes"},
          inplace = True)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Fill the null values:-

```python
df["Killed"] = df["Killed"].fillna(0)
df["Wounded"] = df["Wounded"].fillna(0)
df["Casualty"] = df["Killed"] + df["Wounded"]
```

```python
df.describe()
```

# Visualize the data:-

```python
attacks = df["Year"].value_counts(dropna = False).sort_index().to_frame().reset_index().rename(columns = {"index":"Year","Year":"Attacks"}).set_index("Year")
```

```python
attacks.head()
```

```python
attacks.plot(kind = "bar",color = "green",figsize = (16,9),fontsize = 12)
plt.title("Timeline of Attacks",fontsize = 15)
plt.xlabel("Years",fontsize = 15)
plt.ylabel("Number of Attacks",fontsize = 15)
plt.show()
```

```python
yk = df[["Year","Killed"]].groupby("Year").sum()
yk.head()
```

```python
yk.plot(kind = "bar",color = "orange",figsize = (16,9),fontsize = 12)
plt.title("Timeline of Killed",fontsize = 15)
plt.xlabel("Years",fontsize = 15)
plt.ylabel("Number of Killed",fontsize = 15)
plt.show()
```

```python
yc = df[["Year","Casualty"]].groupby("Year").sum()
yc.head()
```

```python
yc.plot(kind = "bar",color = "m",figsize = (16,9),fontsize = 12)
plt.title("Timeline of Casualty",fontsize = 15)
plt.xlabel("Years",fontsize = 15)
plt.ylabel("Number of Casualty",fontsize = 15)
plt.show()
```

```python
yw = df[["Year","Wounded"]].groupby("Year").sum()
yw.head()
```

```python
yw.plot(kind = "bar",color = "b",figsize = (16,9),fontsize = 12)
plt.title("Timeline of Wounded",fontsize = 15)
plt.xlabel("Years",fontsize = 15)
plt.ylabel("Number of Wounded",fontsize = 15)
plt.show()
```

# Region wise Attacks:-

```python
reg = pd.crosstab(df.Year,df.Region)
reg.head()
```

```python
reg.plot(kind = "area",stacked = False,alpha = 0.5,figsize = (20,10))
plt.title("Region wise Attacks",fontsize = 20)
plt.xlabel("Years",fontsize = 15)
plt.ylabel("Number of Attacks",fontsize = 15)
plt.show()
```

```python
regt = reg.transpose()
regt["Total"] = regt.sum(axis = 1)
rt = regt["Total"].sort_values(ascending = False)
rt
```

```python
rt.plot(kind = "bar",figsize = (16,9))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region",fontsize = 15)
plt.ylabel("Number of Attacks",fontsize = 15)
plt.show()
```
